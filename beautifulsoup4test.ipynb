```python
half_year_pattern = re.compile(r'(\d{2,4}年?)?(上|下)半年|(1|2)H\d{2,4}|\d{2,4}H(1|2)',flags=re.I)
year_pattern = re.compile(r'\d{2,4}年|今年|去年|前年')
month_pattern = re.compile(r'\d{1,2}[\w\-~]\d{1,2}月|\d{1,2}月')
number_pattern = re.compile(r'\d+(\.\d*)?([\w\-~]\d+(\.\d*)?)?(万|亿)?')
per_pattern = re.compile(r'\d+(\.\d*)?\w?(百分点|%|pct|percent)')


# need to identify all stocks first, including contents in the parenthesis
stocks_patterns = 


""" examples
PE 为 6-7 倍


entity
S&P500
"""

def remove_white_spaces(text):
    return re.sub('[ \t\r\f\v]+','',text)

def normalize_year(text):
    return year_pattern.sub('YEAR ',text)

def normalize_half_year(text):
    return half_year_pattern.sub('YEARHALF ',text)


def normalize_month(text):
    return month_pattern.sub('MONTH ',text)

def normalize_percent(text):
    return per_pattern.sub('PERCENT ',text)

def normalize_number(text):
    return number_pattern.sub('NUMBER ',text)

def pipeline(t):
    t = remove_white_spaces(t)
    t = normalize_half_year(t)
    t = normalize_year(t)
    t = normalize_month(t)
    t = normalize_percent(t)
    t = normalize_number(t)
    return t
```

```python
def scrape_industry(save2file='./all_hy.txt'):
    with open(save2file, 'w') as fopen:
        try:
            url = '''http://nufm.dfcfw.com/EM_Finance2014NumericApplication/JS.aspx?type=CT&cmd=C._BKHY&sty=FPGBKI\
&sortType=(Name)&sortRule=1&page=1&pageSize=1000&js={"rank":[(x)]}\
&token=7bc05d0d4c3c22ef9fca8c2a912d779c&jsName=quote_123&_g=0.628606915911589&_=1525970262247'''
        

            try:
                f_url = urlopen(url)
                fp = f_url.read()
            except:
                print('Can not extract {}'.format(url))

            text = None
            try: 
                text = fp.decode('utf-8')
            except UnicodeDecodeError as e:
                text = fp.decode('gb2312')
                print("Decode by GB2312: extract_hy_from_eastmoney({})".format(url))
            except UnicodeDecodeError as e:
                text = fp.decode('gbk')
                print("Decode by GBK: extract_hy_from_eastmoney({})".format(url))
            except:
                print("unknown Error: extract_hy_from_eastmoney({})".format(url))
                return

            if text is None:
                return 

            hy = json.loads(text)['rank']

            #TODO: use sql later

            for h in hy:
                x = h + '\n'
                fopen.write(x)

            print ("Successfully write hy into disk")
        except:
            print ("Fail to write hy into disk")
```
